```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x[1], y[1]
```

```python
plt.figure(figsize=(5,2))
plt.scatter(x,y, c="#FF5733", alpha=.5,marker="*")
plt.xlabel("X -axis")
plt.ylabel("Y-axis")
plt.title("Title : X vs Y")
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
plt.figure(figsize=(9,3))
plt.plot(x,y, "--r")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("X vs Y")
plt.grid()
```

```python
x =["a","b","c","d","e"]
y = np.random.rand(5)
plt.figure(figsize=(5,3))
plt.bar(x,y, color="#FF5733")
plt.xlabel("Catagorical X-axis")
plt.ylabel("Numerical  y-axts")
plt.title("Bar Plot")
plt.grid()
```

```python
y.sort()
plt.figure(figsize=(5,3))
plt.barh(x,y, color="#FF5733")
plt.xlabel("Catagorical X-axis")
plt.ylabel("Numerical  y-axts")
plt.title("Bar Plot")
plt.grid()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
# 3D plot
fig= plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(x,y,z, color="#FF5733", alpha=0.9)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show(fig)
```
