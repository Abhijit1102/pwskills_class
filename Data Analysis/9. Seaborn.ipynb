```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x=tips["total_bill"],y=tips["tip"])
```

```python
tips["smoker"].value_counts()
```

```python
tips
```

```python
sns.relplot(x=tips["total_bill"],y=tips["tip"], hue=tips["smoker"])
```

```python
sns.relplot(x=tips["total_bill"],y=tips["tip"], style=tips["size"])
```

```python
sns.relplot(x=tips["total_bill"],y=tips["tip"], style=tips["size"], hue=tips["time"])
```

```python
sns.catplot(x=tips["day"],y=tips["total_bill"])
```

```python
sns.catplot(x=tips["sex"],y=tips["total_bill"])
```

```python
sns.jointplot(x="smoker",y="tip", data=tips)
```
