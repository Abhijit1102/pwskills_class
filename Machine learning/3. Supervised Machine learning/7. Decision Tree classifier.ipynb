## Decision Tree Algorithm 
A decision tree is a supervised machine learning algorithm used for classification and regression tasks. It creates a model that predicts the value of a target variable based on several input features. The decision tree algorithm partitions the data into smaller subsets by making decisions on the features, ultimately creating a tree-like structure.

Step-by-Step Explanation:

1. `Choosing the Root Node:`
The first step is to choose the root node. It is selected based on a criterion, such as the Gini index or information gain. The root node represents the entire dataset.

2. `Selecting a Splitting Attribute:`
The decision tree algorithm chooses the best attribute to split the data. It evaluates different attributes based on the criterion chosen and selects the one that provides the best split.

In our example, let's assume we use the Gini index as the criterion to measure the impurity of the data. The attribute that yields the lowest Gini index is chosen for splitting.

3. `Splitting the Data:`
Once the splitting attribute is selected, the dataset is divided into subsets based on its possible attribute values. Each subset corresponds to a different branch of the decision tree.

4. `Repeating the Process:`
The process is recursively repeated on each subset, considering the remaining attributes, until the decision tree is fully built.


5. `Stopping Criteria: `
The recursion stops when one of the following conditions is met:

- All instances in a subset belong to the same class (pure node).
- No more attributes are left to split on.
- A predefined maximum depth is reached.
- A minimum number of instances in a subset is reached.
6. `Classification:`
Once the decision tree is built, we can use it to classify new instances. Starting from the root node, we follow the appropriate branch based on the attribute values of the instance until we reach a leaf node. The leaf node represents the predicted class.


Suppose we have a dataset of patients and we want to build a decision tree classifier to predict whether a patient has a specific disease based on their symptoms. The dataset contains the following features: age (in years), gender (male or female), and symptoms (fever, cough, headache)

```python
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import graphviz

# Create the feature matrix X and target vector y
X = [
    ['Sunny', 'Hot', 'High', 'Weak'],
    ['Sunny', 'Hot', 'High', 'Strong'],
    ['Overcast', 'Hot', 'High', 'Weak'],
    ['Rainy', 'Mild', 'High', 'Weak'],
    ['Rainy', 'Cool', 'Normal', 'Weak'],
    ['Rainy', 'Cool', 'Normal', 'Strong'],
    ['Overcast', 'Cool', 'Normal', 'Strong'],
    ['Sunny', 'Mild', 'High', 'Weak'],
    ['Sunny', 'Cool', 'Normal', 'Weak'],
    ['Rainy', 'Mild', 'Normal', 'Weak'],
    ['Sunny', 'Mild', 'Normal', 'Strong'],
    ['Overcast', 'Mild', 'High', 'Strong'],
    ['Overcast', 'Hot', 'Normal', 'Weak'],
    ['Rainy', 'Mild', 'High', 'Strong']
]

y = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']

# Convert categorical features to numerical using one-hot encoding
enc = OneHotEncoder(handle_unknown='ignore')
X_encoded = enc.fit_transform(X).toarray()

# Create a decision tree classifier
clf = tree.DecisionTreeClassifier()

# Train the classifier on the dataset
clf.fit(X_encoded, y)

# Visualize the decision tree
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=enc.get_feature_names_out())
#graph = graphviz.Source(dot_data)
```

```python
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Create the feature matrix X and target vector y
X = [
    ['Sunny', 'Hot', 'High', 'Weak'],
    ['Sunny', 'Hot', 'High', 'Strong'],
    ['Overcast', 'Hot', 'High', 'Weak'],
    ['Rainy', 'Mild', 'High', 'Weak'],
    ['Rainy', 'Cool', 'Normal', 'Weak'],
    ['Rainy', 'Cool', 'Normal', 'Strong'],
    ['Overcast', 'Cool', 'Normal', 'Strong'],
    ['Sunny', 'Mild', 'High', 'Weak'],
    ['Sunny', 'Cool', 'Normal', 'Weak'],
    ['Rainy', 'Mild', 'Normal', 'Weak'],
    ['Sunny', 'Mild', 'Normal', 'Strong'],
    ['Overcast', 'Mild', 'High', 'Strong'],
    ['Overcast', 'Hot', 'Normal', 'Weak'],
    ['Rainy', 'Mild', 'High', 'Strong']
]

y = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']

# Convert categorical features to numerical using one-hot encoding
enc = OneHotEncoder(handle_unknown='ignore')
X_encoded = enc.fit_transform(X).toarray()

# Create a decision tree classifier
clf = tree.DecisionTreeClassifier()

# Train the classifier on the dataset
clf.fit(X_encoded, y)

# Get predicted probabilities for each class
probabilities = clf.predict_proba(X_encoded)

# Compute the false positive rate (FPR), true positive rate (TPR), and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y, probabilities[:, 1], pos_label='Yes')

# Compute the AUC
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

## Implementatin of decession Tree classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
dataset.DESCR
```

```python
dataset.target
```

```python
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
X.head()
```

```python
y = dataset.target
```

```python
# Feature Scaling is no use in tre based algorithm
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42 
)
```

```python
## Post pronnig
from sklearn.tree import DecisionTreeClassifier
classifier =  DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)
```

```python
## Post prunning 
classifier = DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier, filled=True)
```

```python
# Prediction
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred, y_test)
print(score)
print(classification_report(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a decision tree classifier
tree = DecisionTreeClassifier()

# Define the hyperparameters and their possible values for tuning
parameters = {
    'max_depth': [None, 5, 10, 15],
    'splitter': ['best', 'random'],
    'max_features': ['auto','sqrt','log2'],
    'criterion': ['gini', 'entropy']
}

# Perform grid search using cross-validation to find the best hyperparameters
grid_search = GridSearchCV(tree, parameters, cv=5)
grid_search.fit(X, y)

# Get the best hyperparameters and the corresponding score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Train a decision tree classifier using the best hyperparameters
best_tree = DecisionTreeClassifier(**best_params)
best_tree.fit(X, y)
```
